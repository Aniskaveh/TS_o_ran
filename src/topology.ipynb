```python
#---------------------------------------------------------------------------------------------------------------------------
# topology function of "DRL-based Intelligent TS for eMBB and uRLLC coexistance in 6G ORAN", where on RU
# is located in the center of a cell covered by one DU and is divided into three sections covered by one RU per each section
# The users are distributed randomly and equal per each section

# Author: fatemeh.kavehmadavani@uni.lu
# last updated on 2022 Agu 23
# Python version: 3.9.7
#---------------------------------------------------------------------------------------------------------------------------
def topology(U_ur, U_em):

    np.random.seed(1)
    '''x= [0, 350, 0, -300, -350, 180, 400]
    y= [0,200, 400, 200, -240, -400, -10]'''
    x= [0, 0, -350, 180]
    y= [0,400, -240, -400]
    r_outdoor= 500              #Outside radious for RU (meter)
    r_indoor= 100               #Inside radious for RU (meter)
    n= 3                        #number of sectors

    teta= np.arange(0, 2*math.pi, 0.01)
    x1= r_outdoor* np.cos(teta)+ x[0]
    y1= r_outdoor* np.sin(teta)+ y[0]
    x2= r_indoor* np.cos(teta)+ x[0]
    y2= r_indoor* np.sin(teta)+ x[0]
    plt.figure(figsize=(4, 4))
    plt.scatter(x[0], y[0], marker="s", color= 'orange')
    plt.plot(x1, y1) 
    plt.plot(x2, y2) 
    tet= np.linspace(0,2*math.pi,n+1)
    xi= r_outdoor* np.cos(tet)+x[0]
    yi= r_outdoor* np.sin(tet)+y[0]

    #-------------------------divide the circle to n sectors----------------------
    for i in range(len(xi)):
        plt.plot([x[0], xi[i]], [y[0], yi[i]], color= 'black', linestyle = '--')

    plt.scatter(x, y, marker="s", color= 'orange')
    #--------------------------randomly user distribution-------------------------
    n_em, n_ur = int(U_em/n), int(U_ur/n)
    D_em, D_ur = np.zeros((M,U_em)), np.zeros((M,U_ur))

    for i in range(n):
        d_em = (r_outdoor-r_indoor)* np.random.random(n_em)+ r_indoor
        d_ur = (r_outdoor-r_indoor)* np.random.random(n_ur)+ r_indoor
        t_em = (tet[i+1] - tet[i]) * np.random.random(n_em) + tet[i]
        t_ur = (tet[i+1] - tet[i]) * np.random.random(n_ur) + tet[i]
        x_em = x[0] + d_em* np.cos(t_em)  #the x coordinate of int(U_em/n) user at sector i
        y_em = y[0] + d_em* np.sin(t_em)  #the y coordinate of int(U_em/n) user at sector i
        x_ur = x[0] + d_ur* np.cos(t_ur)
        y_ur = y[0] + d_ur* np.sin(t_ur)
        plt.scatter(x_em, y_em, color= 'green', marker="*")
        plt.scatter(x_ur, y_ur, color= 'red', marker="*")
        for m in range(M):
            D_em[m,i*n_em:(i*n_em)+n_em] = np.sqrt((x_em-x[m])**2+(y_em-y[m])**2)
            D_ur[m,i*n_ur:(i*n_ur)+n_ur] = np.sqrt((x_ur-x[m])**2+(y_ur-y[m])**2) 

    dis_em = D_em
    dis_ur = D_ur
    return dis_ur, dis_em
```
